```python
# --- Hromadn√© LLM vyhodnocen√≠ validn√≠ch call≈Ø pro ka≈ædou ot√°zku (OPRAVENO) ---
import glob
import json
import google.generativeai as genai
import os
import time

genai.configure(api_key="AIzaSyB3LkpREqIq8WwCFxsjXEd6-2h-Jnu1G7U")
# Naƒçti v≈°echny summary a p≈ôiprav dictionary {call_name: summary_text}
summary_folder = "./summary"
summary_files = glob.glob(os.path.join(summary_folder, "*_summary.txt"))
all_summaries = {}
for summary_path in summary_files:
    call_name = os.path.basename(summary_path).replace("_summary.txt", "")
    with open(summary_path, "r", encoding="utf-8") as file:
        all_summaries[call_name] = file.read()

def build_llm_prompt_call_list(summaries_dict, question):
    prompt = (
        "You are an expert at finding any possibly relevant project calls.\n"
        "For each CALL below, you will:\n"
        "  1) Think *step by step* whether there is the slightest potential that the answer to the question could be found in its summary.\n"
        "  2) If there is any potential relevance or even the smallest hint or uncertainty, mark it as INCLUDE. Otherwise, mark it as EXCLUDE.\n\n"
        "It is more important to take all pontecial calls. If you are not sure, take that call. \n\n"
        "**** OUTPUT FORMAT ****\n"
        "Return **only** a JSON object where keys are CALL names and values are ‚ÄúINCLUDE‚Äù or ‚ÄúEXCLUDE‚Äù, e.g.:\n"
        "  {\"CALL1\": \"INCLUDE\", \"CALL2\": \"EXCLUDE\", ‚Ä¶}\n\n"
        "Then, in a second JSON list, return **only** the names of the INCLUDED calls, e.g.:\n"
        "  [\"CALL1\", \"CALL5\", ‚Ä¶]\n\n"
        "**** DO NOT output any other text or your chain of thought. ****\n\n"
        f"Question:\n{question}\n\n"
        "Summaries:\n"
    )
    for call, summary in summaries_dict.items():
        prompt += f"CALL: {call}\nSUMMARY: {summary}\n\n"
    return prompt



def generate_call_list(prompt, max_retries=5):
    model = genai.GenerativeModel("gemini-2.5-flash-preview-04-17")
    retries = 0
    while retries < max_retries:
        try:
            response = model.generate_content(prompt)
            # Najdi prvn√≠ a posledn√≠ hranatou z√°vorku a zkus to zparsovat jako list
            json_start = response.text.find('[')
            json_end = response.text.rfind(']') + 1
            json_str = response.text[json_start:json_end]
            parsed = json.loads(json_str)
            return parsed
        except Exception as e:
            retries += 1
            wait_time = 2 ** retries
            print(f"Retry {retries}/{max_retries} after error: {e}. Waiting {wait_time} seconds.")
            time.sleep(wait_time)
    print("Max retries reached. Returning None.")
    return None

llm_summary_answers = [[] for _ in range(len(questions))]

for q_idx, question in enumerate(questions):
    prompt = build_llm_prompt_call_list(all_summaries, question)
    response = generate_call_list(prompt, max_retries=3)
    print(f"LLM response for question {q_idx+1}: {response}")
    if response:
        llm_summary_answers[q_idx] = [f"{call}_combined_text.txt" for call in response]

from sklearn.metrics import precision_score, recall_score, f1_score

# Vyhodnocen√≠ precision, recall, F1 pro ka≈ædou ot√°zku
for q_idx, question in enumerate(questions):
    y_true = []
    y_pred = []
    all_calls = set(answers[q_idx]) | set(llm_summary_answers[q_idx])
    for call in all_calls:
        y_true.append(1 if call in answers[q_idx] else 0)
        y_pred.append(1 if call in llm_summary_answers[q_idx] else 0)
    if len(set(y_true)) > 1:  # mus√≠ b√Ωt aspo≈à jeden true a jeden false
        precision = precision_score(y_true, y_pred)
        recall = recall_score(y_true, y_pred)
        f1 = f1_score(y_true, y_pred)
    else:
        precision = recall = f1 = float('nan')
    print(f"Ot√°zka {q_idx+1}: {question}")
    print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1: {f1:.2f}\n")
```

```python
# --- Vektorizace a roz≈°√≠≈ôen√≠ v√Ωbƒõru call≈Ø podle embedding≈Ø + vyhodnocen√≠ metrik ---
import glob
import os
import torch
from tqdm import tqdm
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics import precision_score, recall_score, f1_score

# 1. Naƒçti v≈°echny combined_text soubory a rozdƒõl na chunk(y)
data_folder = './data'
combined_files = glob.glob(os.path.join(data_folder, '*_combined_text.txt'))
chunk_size = 512  # znak≈Ø na chunk, lze upravit
chunks = []  # [(call_name, chunk_text)]
for file_path in tqdm(combined_files):
    call_name = os.path.basename(file_path).replace('_combined_text.txt', '')
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()
    for i in range(0, len(text), chunk_size):
        chunk = text[i:i+chunk_size]
        chunks.append((call_name, chunk))

# 2. Vygeneruj embeddingy pro v≈°echny chunk(y) pomoc√≠ intfloat/multilingual-e5-base na GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = SentenceTransformer('intfloat/multilingual-e5-base', device=device)
chunk_texts = [c[1] for c in chunks]
chunk_embeddings = model.encode(
    chunk_texts, batch_size=32, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True, device=device
)

# 3. Vygeneruj embedding pro ot√°zku
question_embedding = model.encode([question], convert_to_tensor=True, normalize_embeddings=True, device=device)[0]

# 4. Najdi 10 nejbli≈æ≈°√≠ch chunk≈Ø
cos_scores = util.pytorch_cos_sim(question_embedding, chunk_embeddings)[0]
top_k = min(10, len(chunks))
top_results = cos_scores.topk(top_k)

# 5. Z√≠skej n√°zvy call≈Ø tƒõchto chunk≈Ø
nearest_calls = set()
for idx in top_results.indices.tolist():
    nearest_calls.add(chunks[idx][0])

print('10 nejbli≈æ≈°√≠ch call≈Ø podle embedding≈Ø:', nearest_calls)

# 6. P≈ôidej tyto call names do seznamu doporuƒçen√Ωch call≈Ø (pokud u≈æ tam nejsou)
# Odpovƒõdi z LLM jsou ji≈æ v answers
llm_calls = set([c.replace('_combined_text.txt', '') for c in answers[q_idx]])
all_recommended_calls = llm_calls | nearest_calls
print('Kombinovan√Ω v√Ωbƒõr call≈Ø:', all_recommended_calls)

# 7. Vyhodnocen√≠ metrik pro kombinovan√Ω v√Ωbƒõr
y_true = []
y_pred = []
all_calls = set(answers[q_idx]) | set([f'{c}_combined_text.txt' for c in all_recommended_calls])
for call in all_calls:
    y_true.append(1 if call in answers[q_idx] else 0)
    y_pred.append(1 if call.replace('_combined_text.txt','') in all_recommended_calls else 0)
if len(set(y_true)) > 1:
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
else:
    precision = recall = f1 = float('nan')
print(f'Kombinovan√Ω v√Ωbƒõr - Precision: {precision:.2f}, Recall: {recall:.2f}, F1: {f1:.2f}')

# Pokud chce≈°, m≈Ø≈æe≈° vytvo≈ôit nov√Ω seznam soubor≈Ø pro dal≈°√≠ pou≈æit√≠:
combined_llm_rag_answers = [f'{call}_combined_text.txt' for call in all_recommended_calls]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Convert embeddings to NumPy (from CPU)
qe = question_embeddings.cpu().numpy()   # [num_questions, emb_dim]
ce = chunk_embeddings.cpu().numpy()      # [num_chunks, emb_dim]

num_questions = len(qe)
num_calls = 64  # total number of possible calls

k_values = list(range(1, 301))
precisions = []
recalls = []
false_positive_perc = []  # % of wrongly selected calls

for k in tqdm(k_values, desc="Iterating k"):
    tp = 0        # total true positives
    total_true = 0
    total_pred = 0

    for idx, q_emb in enumerate(qe):
        sims = ce @ q_emb
        topk_idx = np.argpartition(-sims, k-1)[:k]
        nearest_calls = {chunks[i][0] + '_combined_text.txt' for i in topk_idx}

        llm_calls = {name + '_combined_text.txt' for name in llm_responses[idx]}
        true_calls = set(true_calls_list[idx])

        pred_calls = nearest_calls | llm_calls
        correct = true_calls & pred_calls

        tp += len(correct)
        total_true += len(true_calls)
        total_pred += len(pred_calls)

    fp_total = total_pred - tp  # all false predictions
    precisions.append(tp / total_pred if total_pred > 0 else np.nan)
    recalls.append(tp / total_true if total_true > 0 else np.nan)
    # average wrong calls per question = fp_total/num_questions
    # as percentage of all possible calls:
    false_positive_perc.append((fp_total / num_questions) / num_calls * 100)

# plot Precision & Recall
plt.figure()
plt.plot(k_values, recalls, label="Recall")
plt.plot(k_values, precisions, label="Precision")
plt.xlabel("k (number of nearest RAG calls)")
plt.ylabel("Micro metric")
plt.title("Precision & Recall for RAG+LLM vs. k")
plt.legend()
plt.grid(True)
plt.show()

# plot % of wrong calls
plt.figure()
plt.plot(k_values, false_positive_perc)
plt.xlabel("k (number of nearest RAG calls)")
plt.ylabel("Average % of incorrectly selected calls")
plt.title("Average percentage of bad calls vs. k")
plt.grid(True)
plt.show()
```
