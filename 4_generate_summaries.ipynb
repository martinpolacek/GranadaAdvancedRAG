```python
# Load valid projects from valid_projects.txt
valid_projects = []
with open('valid_project.txt', 'r') as file:
    for line in file:
        project = line.strip()
        if project:  # Ignore empty lines
            valid_projects.append(project)
```

```python
import os
import google.generativeai as genai
from google.generativeai import types
import time  # Added for retry logic

# Nastav API klíč (ujisti se, že máš ve svém prostředí proměnnou GOOGLE_API_KEY)
genai.configure(api_key="AIzaSyB3LkpREqIq8WwCFxsjXEd6-2h-Jnu1G7U")


def generate(input: str, max_retries=5):
    model = genai.GenerativeModel("gemini-2.0-flash")
    prompt = """
    Provide a comprehensive, well-structured summary that must include the following key information:
    
    1. WHO: Who is organizing the project and who are the eligible applicants or target groups
    2. WHAT: What type of program or call it is and what is the exact focus of the project
    3. HOW: What is the procedure for submitting applications and key requirements for participants
    4. WHY: What are the objectives and expected results of the project
    5. HOW MUCH: What is the amount of support, financial scope of projects, co-financing rate
    6. OTHER IMPORTANT INFORMATION: Deadlines, specific conditions, limitations, and other essential details
    
    Data:
    
    """
    prompt = prompt + input + "\n\n"
    retries = 0
    while retries < max_retries:
        try:
            response = model.generate_content(prompt)
            # Najdi první a poslední složenou závorku a zkus to zparsovat
            return response.text
        except Exception as e:
            retries += 1
            wait_time = 2 ** retries  # Exponential backoff
            print(f"Retry {retries}/{max_retries} after error: {e}. Waiting {wait_time} seconds.")
            time.sleep(wait_time)
    print("Max retries reached. Returning None.")
    return None
```

```python
#
SUMMARY_DIR = "summary"
DATA_DIR = "data"

for project in valid_projects:

    print(f"Processing project: {project}")



    data_file = os.path.join(DATA_DIR, f"{project}_combined_text.txt")
    summary_file = os.path.join(SUMMARY_DIR, f"{project}_summary.txt")

    #If summary file already exists, skip
    if os.path.exists(summary_file):
        print(f"Summary file {summary_file} already exists. Skipping.")
        continue

    # Zkontroluj, zda existuje soubor s daty
    if not os.path.exists(data_file):
        print(f"File {data_file} does not exist. Skipping.")
        continue
    # Načti data ze souboru
    with open(data_file, 'r') as file:
        data = file.read()

    #Generuj shrnutí
    summary = generate(data)
    if summary is None:
        print(f"Failed to generate summary for {project}.")
        continue
    # Ulož shrnutí do souboru and create file if it doesn't exist
    with open(summary_file, 'w') as file:
        file.write(summary)
    print(f"Summary for {project} saved to {summary_file}.")
```
